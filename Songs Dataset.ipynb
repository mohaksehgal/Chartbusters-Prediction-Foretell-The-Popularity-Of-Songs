```python
def u(x):
    n=''
    for i in str(x):
        if i==',':
            continue
        else:
            n=n+i
    return(n)
```

```python
def u1(x):
    n=str(x)
    if n[-1]=='K':
        n1=''
        for i in n:
            if i=='.' or i=='K':
                continue
            else:
                n1=n1+i
        w=n1+'00'
        return(int(w))
    else:
        return(n)
```

```python
def u2(x):
    n=str(x)
    if n[-1]=='M':
        n1=''
        for i in n:
            if i=='.' or i=='M':
                continue
            else:
                n1=n1+i
        w=n1+'000000'
        return(int(w))
    else:
        return(int(n))
```
