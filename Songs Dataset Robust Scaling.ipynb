```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
A=pd.read_excel('Cleaned File.xlsx')
```

```python
A.head()
```

```python
A_dummy=pd.get_dummies(A,drop_first=True)
```

```python
A_dummy.head()
```

```python
Train=A_dummy[A_dummy['S']==0]
```

```python
Test=A_dummy[A_dummy['S']==1]
```

```python
Train.drop('S',axis=1,inplace=True)
Test.drop(['S','Views'], axis=1,inplace=True)
```

```python
Train.shape
```

```python
Test.shape
```

```python
Train.head()
```

```python
Test.head()
```

```python
Train.drop('Unique_ID',axis=1,inplace=True)
Test.drop('Unique_ID',axis=1,inplace=True)
```

```python
Train.head()
```

```python
Test.head()
```

```python
Train=Train.reset_index(drop=True)
Test=Test.reset_index(drop=True)
```

```python
Train.head()
```

```python
Test.head()
```

```python
Train.shape
```

```python
Test.shape
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
RS=RobustScaler()
```

```python
x=Train.drop('Views',axis=1)
y=Train['Views']
```

```python
x_scaled=RS.fit_transform(x)
Test_scaled=RS.transform(Test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()
```

```python
ypre=LR.fit(x_scaled,y).predict(Test_scaled)
```

```python
sample=pd.read_excel('Sample_Submission.xlsx')
```

```python
sample['Views']=ypre
```

```python
sample.to_excel('LR(RS).xlsx',index=False)
```
